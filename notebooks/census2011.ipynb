### 1. Number of records by hours worked per week and industry

```python
# 替换编码为描述符的函数
def map_labels(column, mapping):
    return column.map(mapping).fillna(column)

# 计算按每周工作时数和行业划分的记录数量
grouped_data = data.groupby(['hours_worked_per_week', 'industry']).size().reset_index(name='record_count')
grouped_data['hours_worked_per_week'] = map_labels(grouped_data['hours_worked_per_week'], labels['Hours_Worked_Per_Week'])
grouped_data['industry'] = map_labels(grouped_data['industry'], labels['industry'])

# 绘制分组条形图
plt.figure(figsize=(12, 8))
for hour_category in grouped_data['hours_worked_per_week'].unique():
    subset = grouped_data[grouped_data['hours_worked_per_week'] == hour_category]
    plt.bar(subset['industry'], subset['record_count'], label=hour_category)

# 添加图表信息
plt.title('Number of Records by Hours Worked Per Week and Industry', fontsize=16)
plt.xlabel('Industry', fontsize=14)
plt.ylabel('Number of Records', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Hours Worked Per Week')
plt.tight_layout()  # 调整布局以防止标签被遮挡

# 显示图表
plt.show()
```

### 2.Number of records by occupation and approximate social grade

```python
# 替换编码为描述符的函数
def map_labels(column, mapping):
    return column.map(mapping).fillna(column)

# 计算按职业和大致社会等级划分的记录数量
grouped_data = data.groupby(['occupation', 'approximate_social_grade']).size().reset_index(name='record_count')
grouped_data['occupation'] = map_labels(grouped_data['occupation'], labels['Occupation'])
grouped_data['approximate_social_grade'] = grouped_data['approximate_social_grade'].astype(str)

# 绘制分组条形图
plt.figure(figsize=(14, 8))
for grade in grouped_data['approximate_social_grade'].unique():
    subset = grouped_data[grouped_data['approximate_social_grade'] == grade]
    plt.bar(subset['occupation'], subset['record_count'], label=f'Social Grade {grade}')

# 添加图表信息
plt.title('Number of Records by Occupation and Approximate Social Grade', fontsize=16)
plt.xlabel('Occupation', fontsize=14)
plt.ylabel('Number of Records', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Approximate Social Grade')
plt.tight_layout()  # 调整布局以防止标签被遮挡

# 显示图表
plt.show()
```

### Interactively control plot 

```python
# 提取“health”和“Economic_Activity”映射字典
health_mapping = labels['health']
economic_activity_mapping = labels['Economic_Activity']

# 将CSV数据中的编码值替换为描述性文本
data['health'] = data['health'].replace(health_mapping)
data['economic_activity'] = data['economic_activity'].replace(economic_activity_mapping)

# 3. 过滤出经济不活跃的群体（退休，学生等）
inactive_data = data[data['economic_activity'].str.contains("Economically inactive")]

# 4. 创建交互式控件
health_dropdown = widgets.Dropdown(
    options=['All'] + list(health_mapping.values()),  # 选项包括所有健康描述符
    description='Health:',
)

age_slider = widgets.IntSlider(
    value=15,
    min=0,
    max=75,
    step=5,
    description='Age:',
    continuous_update=False
)

# 5. 定义更新图表的函数
def update_plot(health, age):
    # 如果选择了“所有”，则忽略筛选健康条件
    if health != 'All':
        filtered_data = inactive_data[(inactive_data['health'] == health) & (inactive_data['age'] == age)]
    else:
        filtered_data = inactive_data[inactive_data['age'] == age]
    
    # 统计健康描述符和年龄组合的数量
    health_counts = filtered_data['health'].value_counts()
    
    # 绘制条形图
    plt.figure(figsize=(10, 6))
    health_counts.sort_index().plot(kind='bar', color='lightblue', edgecolor='black')
    
    # 设置标题和标签
    plt.title(f"Number of Economically Inactive People by Health Descriptor (Age: {age})", fontsize=16)
    plt.xlabel('Health Descriptor', fontsize=14)
    plt.ylabel('Number of People', fontsize=14)
    
    # 在柱子上添加数量标签
    for i, value in enumerate(health_counts):
        plt.text(i, value + 0.1, str(value), ha='center', fontsize=12)  # 在柱子上方显示数量
    
    # 显示图表
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# 6. 使用ipywidgets的交互式功能，控制图表更新
widgets.interactive(update_plot, health=health_dropdown, age=age_slider)

# 显示控件
from IPython.display import display
display(health_dropdown, age_slider)
```
